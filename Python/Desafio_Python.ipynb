AGRUPAMENTO DE DADOS E PLOTAGEM DE GRÃFICOS

```python
df_bar = df.groupby('Department')['EmployeeCount'].sum()
df_bar
```

```python
df_bar.plot(kind='bar')
plt.xticks(rotation=0)
```

```python
df_line = df.groupby('Education')['EmployeeCount'].sum()
df_line
```

```python
df_line.plot(kind='line', marker='o')
plt.xticks(rotation=0)
plt.title('NÃ­vel de Escolaridade dos FuncionÃ¡rios')
```

```python
df_pie =  df.groupby('Attrition')['EmployeeCount'].sum()
df_pie
```

```python
labels = ['Attrition', 'EmployeeCount']
sizes = 15, 30
color = ['gold', 'yellowgreen']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=color, autopct='%1.1f%%',
        shadow=True, startangle=140)

plt.axis('equal')  # Assegura que o grÃ¡fico seja desenhado como um cÃ­rculo.
plt.title('Quantidade de FuncionÃ¡rios em Atrito')
plt.show()
```
